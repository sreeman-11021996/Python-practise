```python
import itertools
```

```python
letters = ['a','b','c','d']
numbers = [0, 1, 2, 3]
names = ['Corey', 'Nicole']
```

```python
result = itertools.combinations(letters,2) # oreder does not matter
```

```python
list(result)
```

```python
result_com_rep = itertools.combinations_with_replacement(numbers,4)
```

```python
list(result_com_rep)
```

```python
result_p = itertools.permutations(letters,2) # order does matter
```

```python
list(result_p)
```

```python
result_rep_pd = itertools.product(numbers,numbers,numbers,numbers)
```

```python
list(result_rep_pd)
```
