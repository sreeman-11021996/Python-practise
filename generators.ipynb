```python
def square(n):
    return [i**2 for i in range(n)]    
```

```python
import sys
```

```python
lst = square(1000000)
```

```python
sys.getsizeof(lst)
```

```python
class gen:
    def __init__ (self,n):
        self.n = n
        self.last = 0    # 1st value
    
    def __next__ (self):
        return self.next()
    
    def next(self):
        if self.last == self.n:
            raise StopIteration()
        rv = self.last**2
        self.last+=1
        return rv
```

```python
g = gen(1000000000)
```

```python
sys.getsizeof(g)
```

```python
while True:
    try:
        print(next(g))
    except StopIteration:
        break
```

```python
def square_gen(n):
    for i in range(n):
        yield i**2  # generator keyword - function is paused here
```

```python
g = square_gen(1000)
```

```python
sys.getsizeof(g)
```

```python
for i in g : print(i)
```
