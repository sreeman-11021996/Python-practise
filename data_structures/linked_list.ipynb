```python
class Node:
    
    def __init__ (self,value):
        self.data = value
        self.next = None  #pointer to the next node
        
```

```python
class Linked_list:
    
    def __init__ (self):
        self.head = None
    
    def insert(self,value):     #insert from head
        # create a node 
        new_node = Node(value)
        # point the node to first element / head
        new_node.next = self.head
        #point head to new_node
        self.head = new_node
    
    def insert_after (self, prev_data, new_data):
        #find the prev_node
        prev_node = self.head
        while prev_node is not None:
            if (prev_node.data == prev_data):
                break
            prev_node = prev_node.next
        if(prev_node == None):
            print("Node not found")
        else:
            # Insert the new node
            new_node = Node(new_data)
            new_node.next = prev_node.next
            prev_node.next = new_node
    
    def insert_end (self, value):
        new_node = Node(value)
        #check if linked list is empty
        if self.head is None:
            self.head = new_node
        # traverse the list and insert
        else:
            tail = self.head
            while tail.next is not None:
                tail = tail.next
            tail.next = new_node     
    
    def delete_head (self):
        # If List is Empty
        if self.head is None:
            print("Cannot Delete! Linked List is 'Empty'")
        else:
            self.head = self.head.next
            
    def delete_tail(self):
        # If List is Empty
        if self.head is None:
            print("Cannot Delete! Linked List is 'Empty'")
        elif (self.head.next is None):
            self.head = None
        else:
            #traverse the list
            temp = self.head
            while temp.next.next is not None:
                temp = temp.next
            temp.next = None
            
    def delete_item (self, item):
        # If List is Empty
        if self.head is None:
            print("Cannot Delete! Linked List is 'Empty'")
        # If the item is head
        elif self.head.data == item :
            self.head = self.head.next
        else:
            temp = self.head
            while temp.next is not None:
                if temp.next.data == item :
                    break
                temp = temp.next
            # make the new connection
            if temp.next is None:
                print("Item not found")
            else:
                temp.next = temp.next.next
        
    def traverse (self):
        temp = self.head
        while temp is not None :
            print(temp.data,end = " -> ")
            temp = temp.next
        print("Null")
```
